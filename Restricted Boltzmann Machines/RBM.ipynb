```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.examples.tutorials.mnist import input_data


# input data
mnist = input_data.read_data_sets('MNIST_data', one_hot=True)


# RBM
def run_rbm_model(input_x, n_visible=784, n_hidden=500, momentum=0.95, learning_rate=0.01):
    w = tf.Variable(tf.truncated_normal([n_visible, n_hidden]), dtype=tf.float32)
    a = tf.Variable(tf.zeros(shape=[n_visible, 1]), dtype=tf.float32)
    b = tf.Variable(tf.zeros(shape=[n_hidden, 1]), dtype=tf.float32)
    end_points = {}

    # define sampling method
    def sample_bernoulli(probs):
        return tf.nn.relu(tf.sign(probs - tf.random_uniform(tf.shape(probs))))
    

    # reconstruct variables
    hidden_p = tf.nn.sigmoid(tf.matmul(tf.transpose(w), tf.transpose(input_x)) + b)
    # h = sample_bernoulli(hidden_p)
    h = hidden_p
    visible_reconstruct_p = tf.nn.sigmoid(tf.matmul(w, h) + a)
    # v_reconstruct = sample_bernoulli(visible_reconstruct_p)
    v_reconstruct = visible_reconstruct_p
    hidden_reconstruct_p = tf.nn.sigmoid(tf.matmul(tf.transpose(w), v_reconstruct) + b)
    h_reconstruct = hidden_reconstruct_p

    positive_grad = tf.matmul(tf.transpose(input_x), tf.transpose(hidden_p))
    negative_grad = tf.matmul(v_reconstruct, tf.transpose(h_reconstruct))

    # use momentum learning method
    def f(x_old, x_new):
        return momentum * x_old + \
               learning_rate * x_new * (1 - momentum) / tf.to_float(tf.shape(x_new)[0])

    # update weights
    w = f(w, positive_grad - negative_grad)
    a = f(a, tf.transpose(input_x) - v_reconstruct)
    b = f(b, h - h_reconstruct)

    # compute loss
    hidden_p_new = tf.nn.sigmoid(tf.matmul(tf.transpose(w), tf.transpose(input_x)) + b)
    # h_new = sample_bernoulli(hidden_p_new)
    h_new = hidden_p_new
    visible_reconstruct_p_new = tf.nn.sigmoid(tf.matmul(w, h_new) + a)
    # v_reconstruct_new = sample_bernoulli(visible_reconstruct_p_new)
    v_reconstruct_new = visible_reconstruct_p_new
    compute_loss = tf.reduce_mean(tf.square(tf.transpose(input_x) - v_reconstruct_new))

    end_points['loss'] = compute_loss
    end_points['v_recons'] = v_reconstruct_new

    return end_points


def run_training(epochs, batch_size):
    batch_x = tf.placeholder(tf.float32, [1, 784])
    end_points = run_rbm_model(input_x=batch_x)
    saver = tf.train.Saver()
    init_op = tf.group(tf.global_variables_initializer(), tf.local_variables_initializer())

    with tf.Session() as sess:
        sess.run(init_op)
        for epoch in range(epochs):
            batch_xs, batch_ys = mnist.train.next_batch(batch_size)
            for batch in range(batch_size):
                batch_input = np.reshape(batch_xs[batch], [1, 784])
                loss = sess.run(end_points['loss'], feed_dict={batch_x: batch_input})
                if batch % 10 == 0:
                    print('[INFO] Epoch: %d , batch: %d , training loss: %.6f' % (epoch, batch, loss))
        saver.save(sess, './training_RBM')


def image_generate():
    input_x = tf.placeholder(tf.float32, [1, 784])
    end_points = run_rbm_model(input_x=input_x)

    saver = tf.train.Saver(tf.global_variables())
    init_op = tf.group(tf.global_variables_initializer(), tf.local_variables_initializer())

    def show_image(x):
        plt.imshow(x.reshape((28, 28)), cmap="gray")
        plt.show()

    with tf.Session() as sess:
        sess.run(init_op)
        saver.restore(sess, './training_RBM')
        for i in range(20):
            image = mnist.test.images[i]
            input_image = np.reshape(image, [1, 784])
            image_recon = sess.run(end_points['v_recons'], feed_dict={input_x: input_image})
            show_image(image)
            show_image(image_recon)

# training and generate images
tf.reset_default_graph()
run_training(epochs=600, batch_size=100)
tf.reset_default_graph()
image_generate()

```
